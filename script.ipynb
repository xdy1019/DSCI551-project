```python
stocks = ["AAPL", "AMAT", "MSFT", "AMZN"]
users = list(range(1, 51))  # 50 users

portfolio = {(user, stock): 0 for user in users for stock in stocks}

all_trades = []

for stock in stocks:
    df = yf.download(stock, start="2025-03-01", end="2026-03-01")
    df.columns = df.columns.get_level_values(0)
    df = df.reset_index()
    df["Date"] = pd.to_datetime(df["Date"])

    for _, row in df.iterrows():
        date = row["Date"]
        price = float(row["Close"])

        for _ in range(random.randint(10, 30)):
            user = random.choice(users)
            holding = portfolio[(user, stock)]

            # decide side safely
            if holding == 0:
                side = "Buy"
            else:
                side = random.choices(["Buy", "Sell"], weights=[0.6, 0.4])[0]

            # generate quantity properly
            if side == "Buy":
                quantity = random.randint(1, 50)
                portfolio[(user, stock)] += quantity

            else:  # SELL
                quantity = random.randint(1, holding)  
                portfolio[(user, stock)] -= quantity

            trade = {
                "user_id": user,
                "stock": stock,
                "price": round(price * random.uniform(0.98, 1.02), 2),
                "quantity": quantity,
                "side": side,
                "timestamp": date
            }

            all_trades.append(trade)

trades_df = pd.DataFrame(all_trades)
trades_df.to_csv("data.csv", index=False)
```
